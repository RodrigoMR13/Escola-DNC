```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
vendas = pd.read_csv("Walmart.csv")
```

```python
vendas_feriados = vendas[vendas['Holiday_Flag'] == 1]
```

```python
medias_feriados = vendas_feriados.groupby('Store').mean()
medias_feriados = medias_feriados.sort_values('Weekly_Sales', ascending=False)
medias_feriados.head()
```

```python
medias_das_lojas = vendas.groupby('Store').mean().reset_index()
medias_das_lojas = medias_das_lojas.sort_values('Weekly_Sales', ascending=False)
medias_das_lojas.head()
```

```python
plt.figure(figsize=(17,7))
plt.bar(medias_das_lojas['Store'], medias_das_lojas['Weekly_Sales'])
plt.xticks(np.arange(1,46,1))
plt.xlabel('Número da Loja')
plt.yticks(np.arange(0.1*10**6,2.3*10**6,10**5))
plt.ylabel('Vendas (em milhões)')
plt.title('Média de Vendas das Lojas');
```

```python
soma_das_vendas = vendas.groupby('Store').sum().reset_index()
soma_das_vendas = soma_das_vendas.sort_values('Weekly_Sales', ascending=False)
soma_das_vendas
```

```python
plt.figure(figsize=(17,7))
plt.bar(soma_das_vendas['Store'], soma_das_vendas['Weekly_Sales'])
plt.xticks(np.arange(1,46,1))
plt.xlabel('Número da Loja')
plt.yticks(np.arange(0.1*10**8,3.1*10**8,10**7))
plt.ylabel('Vendas (em milhões)')
plt.title('Soma de Vendas das Lojas');
```

```python
filtro = vendas[vendas['Store']==20]
filtro[vendas['Weekly_Sales']>2107677].count()
```

```python
vendas_max = vendas.groupby('Store').max('Weekly_Sales').reset_index()
vendas_max.sort_values('Weekly_Sales', ascending=False)
```

```python
vendas_min = vendas.groupby('Store').min('Weekly_Sales').reset_index()
vendas_min.head()
```

```python

fig,ax = plt.subplots(figsize=(18,12))

grupo1 = ax.bar(vendas_max['Store']-0.45/2, vendas_max['Weekly_Sales'],width=0.45, label='Venda Máxima', color='teal')
grupo2 = ax.bar(medias_das_lojas['Store']+0.45/2, medias_das_lojas['Weekly_Sales'], width=0.45, label='Média de Vendas', color='indigo')
grupo3 = ax.bar(vendas_min['Store']-0.45/2, vendas_min['Weekly_Sales'], width=0.45, label='Venda Mínima', color='tomato')

ax.set_title('Análise das Vendas')
ax.set_xlabel('Número da Loja')
ax.set_ylabel('Vendas (em milhões)')
ax.set_xticks(np.arange(1,46,1))
ax.set_yticks(np.arange(0.1*10**6, 4.1*10**6, 2*10**5))
ax.legend()

plt.show()
```
