```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv')
df
```

```python
# Em primeiro lugar, tratei de identificar se havia algum tipo de dado que tivesse que ser mudado.
df.info()
```

```python
# Notei que era necess√°rio mudar a data para datetime, os banheiros para inteiros, os andares para inteiros e a condi√ß√£o do im√≥vel ser beira-mar para boolean.
df['date'] = pd.to_datetime(df['date'])
df['bathrooms'] = df['bathrooms'].astype('int64')
df['floors'] = df['floors'].astype('int64')
df['waterfront'] = df['waterfront'].astype('boolean')
df.info()
```

```python
# Em seguida, tratei de eliminar qualquer dado nulo que pudesse estar no meu dataset, mas pelo que vi, n√£o havia nenhum dado nulo.
df = df.dropna()
df
```

```python
# Ap√≥s a elimina√ß√£o dos dados nulos, criei uma nova coluna para identificar o pre√ßo por √°rea de cada im√≥vel.
df['price_per_area'] = df['price'] / df['sqft_lot']
df.head()
```

```python
# Chegando na parte de an√°lise, tratei de identificar os im√≥veis a beira-mar, com grau 4 de beleza da vista e com pre√ßos menores que $500.000,00.
# Al√©m disso, ordenei os im√≥veis por ordem decrescente de √°rea do terreno, dessa forma identifiquei 5 im√≥veis baratos com grande possibilidade de valoriza√ß√£o ap√≥s construir no terreno.
# Apenas dois me chamaram aten√ß√£o: id = 2013802030 e id = 2013802060, os dois vendidos a um pre√ßo baixo, mas com grande √°rea de terreno, pre√ßo por area de aproximadamente 6,62 e 18,29 
# e localizados em Des Moines, uma cidade muito perto do centro de Seattle.
best_sqft_lot = df[(df['waterfront'] == True) & (df['view'] == 4) & (df['price'] <= 500000)].sort_values(by=['sqft_lot'], ascending=False)
best_sqft_lot.head()
```

```python
# Analisando os im√≥veis nas mesmas condi√ß√µes, por√©m ordenando em ordem crescente de pre√ßo, encontrei im√≥veis baratos com √°reas grandes para constru√ß√£o.
# A maioria localizado na ilha de Vashon, mas um me chamou aten√ß√£o, o im√≥vel com id = 7631800110. Esse im√≥vel tamb√©m foi vendido a um pre√ßo baixo, pre√ßo por area de 21,91 e possui
# muita possibilidade de crescimento por conta da localiza√ß√£o (perto do centro de Seattle) e √°rea do terreno.
best_price = df[(df['waterfront'] == True) & (df['view'] == 4) & (df['price'] <= 500000)].sort_values(by=['price'], ascending=True)
best_price.head()
```
