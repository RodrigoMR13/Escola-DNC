```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv')
df
```

```python
# Em primeiro lugar, tratei de identificar se havia algum tipo de dado que tivesse que ser mudado.
df.info()
```

```python
# Notei que era necessário mudar a data para datetime, os banheiros para inteiros, os andares para inteiros e a condição do imóvel ser beira-mar para boolean.
df['date'] = pd.to_datetime(df['date'])
df['bathrooms'] = df['bathrooms'].astype('int64')
df['floors'] = df['floors'].astype('int64')
df['waterfront'] = df['waterfront'].astype('boolean')
df.info()
```

```python
# Em seguida, tratei de eliminar qualquer dado nulo que pudesse estar no meu dataset, mas pelo que vi, não havia nenhum dado nulo.
df = df.dropna()
df
```

```python
# Após a eliminação dos dados nulos, criei uma nova coluna para identificar o preço por área de cada imóvel.
df['price_per_area'] = df['price'] / df['sqft_lot']
df.head()
```

```python
# Chegando na parte de análise, tratei de identificar os imóveis a beira-mar, com grau 4 de beleza da vista e com preços menores que $500.000,00.
# Além disso, ordenei os imóveis por ordem decrescente de área do terreno, dessa forma identifiquei 5 imóveis baratos com grande possibilidade de valorização após construir no terreno.
# Apenas dois me chamaram atenção: id = 2013802030 e id = 2013802060, os dois vendidos a um preço baixo, mas com grande área de terreno, preço por area de aproximadamente 6,62 e 18,29 
# e localizados em Des Moines, uma cidade muito perto do centro de Seattle.
best_sqft_lot = df[(df['waterfront'] == True) & (df['view'] == 4) & (df['price'] <= 500000)].sort_values(by=['sqft_lot'], ascending=False)
best_sqft_lot.head()
```

```python
# Analisando os imóveis nas mesmas condições, porém ordenando em ordem crescente de preço, encontrei imóveis baratos com áreas grandes para construção.
# A maioria localizado na ilha de Vashon, mas um me chamou atenção, o imóvel com id = 7631800110. Esse imóvel também foi vendido a um preço baixo, preço por area de 21,91 e possui
# muita possibilidade de crescimento por conta da localização (perto do centro de Seattle) e área do terreno.
best_price = df[(df['waterfront'] == True) & (df['view'] == 4) & (df['price'] <= 500000)].sort_values(by=['price'], ascending=True)
best_price.head()
```
